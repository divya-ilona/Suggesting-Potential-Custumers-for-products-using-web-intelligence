```python
########### input data for products ################

col_names = ["Title","Title_URL", "Field", "Title_URL1", "Title2", "Title3"] 
# data = pd.read_csv('drive/My Drive/VM_Ware/products.csv',names=col_names)
data = pd.read_csv('/content/drive/My Drive/research/VMWARE_DATA/prod_vm.csv',names=col_names)
data.columns = col_names
data = data.drop([0], axis=0)
print(len(data))
```

```python
s1 = data.Title
s2 = data.Field
s3 = data.Title2
s4 = data.Title3

l =[]
s1 = s1.tolist()
s2 = s2.tolist()
s3 = s3.tolist()
s4 = s4.tolist()
print(len(s1), len(s3), len(s3), len(s4))
tokenized_l =[]
prod_k = {}
word_list = []
s1l = []
s2l = []
s3l = []
s4l = []
for x in s1:
  if str(x) != 'nan':
    s1l.append(x)
  else:
    s1l.append('t')

for x in s2:
  if str(x) != 'nan':
    s2l.append(x)
  else:
    s2l.append('t')
for x in s3:
  if str(x) != 'nan':
    s3l.append(x)
  else:
    s3l.append('t')
for x in s4:
  if str(x) != 'nan':
    s4l.append(x)
  else:
    s4l.append('t')
key_words_list = set()
for i in range(len(s1)):
  vl = []
  m = word_tokenize(s1l[i])
  key_words_list.add(s1l[i])
  vl.extend(m)
  m = word_tokenize(s2l[i])
  key_words_list.add(s1l[i])
  vl.extend(m)
  m = word_tokenize(s3l[i])
  key_words_list.add(s1l[i])
  vl.extend(m)
  m = word_tokenize(s4l[i])
  key_words_list.add(s1l[i])
  vl.extend(m)
  prod_k[s1l[i]] = vl

# print(prod_k)
# break
punctuations=list(punctuation)
stopWords=stopwords.words('english')
stopWords+=punctuations

final_prod = {}
for k,v in prod_k.items():
  final = []
  for w in v:
    if not w.lower() in stopWords:
      final.append(w)
  final_prod[k] = final
  
print(final_prod) ### product to keywords mapping
key_prod = {}
for k,v in final_prod.items():
  # print(k)
  for w in v:
    # print(w)
    if w in key_prod:
      # print(key_prod[w])
      key_prod[w].add(k)
    else:
      l= set()
      l.add(k)
      key_prod[w] = l
print(key_prod) ### keywords to product mapping

print("Length of keywords",len(key_words_list))
print(key_words_list)
```
