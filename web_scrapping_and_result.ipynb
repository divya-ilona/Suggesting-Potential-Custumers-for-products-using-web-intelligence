```python
import os
import pandas as pd
from datetime import date
```

```python
search_term = 'start-up'
from_date = '2021-10-08'
end_date = '2021-10-11'
os.system(f"snscrape --since {from_date} twitter-search '{search_term} until:{end_date}' > result-tweets.txt")
if os.stat("result-tweets.txt").st_size == 0:
    counter = 0
else:
    df1 = pd.read_csv('result-tweets.txt', names=['link'])
    counter = df1.size

print('Number Of Tweets : '+ str(counter))
```

```python
temp_df=df1
url = df1['link']
url.to_list()
print(url)
list1=df1
list1['2'] = df1['link'].str.split("https://twitter.com/").str[1]
list1['username'] = list1['2'].str.split('/').str[0]
df3=list1['username']
df3
```

```python
username1=df1['username']
username1.to_list()
```

```python
max_results = 3000
extracted_tweets = "snscrape --format '{content!r}'"+ f" --max-results {max_results} --since {from_date} twitter-search '{search_term} until:{end_date}' > extracted-tweets.txt"
os.system(extracted_tweets)
if os.stat("extracted-tweets.txt").st_size == 0:
  print('No Tweets found')
else:
  df = pd.read_csv('extracted-tweets.txt', names=['content'])
  for row in df['content'].iteritems():
    print(row)
```

```python
df2=pd.DataFrame(df)
```

```python
list2 = df2
list2 = list2['content'].to_list()
print(list2)
```

```python
keyword = [ "cloud","networking","applications","workspace","storage","security","virtualization", "virtualization software", "network and security virtualization", "Digital workspace platform that enables zero trust", "Kubernetes", "Flash-optimized, vSphere-native storage", "security portfolio", "cloud service","Artificial Intelligence","ML","server","AI",]
necessary_keyword = ["service", "need", "newstartup", "cloud", "require","developing","automate","develop","platform","industry","grow","growing","scaling","scale","scaled","funding","raises","product"]

customer=[]
ind =-1
buying_list=[]
profile_link = []

for i in list2:
  ind=ind+1 
  for j in keyword:
    if i.find(j)>0:
      count=0
      for k in necessary_keyword:
        if i.find(k) > 0:
          count = count+1
      if(count>0):
        name=username1[ind]
        profile = url[ind]
        customer.append(name)
        profile_link.append(profile)
        buying_list.append(j)
        break
```

```python
customer
```

```python
product_list=['AI','VMware Cloud','Tanzu Basic','vSphere']
product_dsp=[['AI','ML'],['cloud','storage','cloud services'],['Kubernetes','multi-cloud Kubernetes'],['server',"server virtualization"]]
suggested_product=[]
cnt=-1
for i in customer:
  cnt=cnt+1
  index=-1
  for j in product_list:
     index=index+1
     for k in product_dsp[index]:
       if(k==buying_list[cnt]):
         suggested_product.append(j)
         break
        
```
